## Time to make some plots

```python
# Get top 10 contries w/ cleanest water
Cleanest = df.sort_values(by = 'Sanitation & Drinking Water', ascending = False)[:10]
print(Cleanest["country_name"].unique())
```

```python
# Size
plt.figure(figsize = (15, 5))

# Bar chart
plt.bar(x = Cleanest['country_name'], height = Cleanest['Sanitation & Drinking Water'])

# Create a descriptive title
plt.title('Top 10 contries w/ cleanest water')

# Add axis labels
plt.xlabel('country_name')
plt.ylabel('Amount')

# Rotate text
plt.xticks(rotation = 10); # semicolon suppresses text output
```

```python
# Size
plt.figure(figsize = (15, 20))

# Line chart
plt.plot(df['GDP'], df['Climate'])

# Create a descriptive title
plt.title('Environmental Performance Index')

# Add axis labels
plt.ylabel('Percentage of People'); 
# X label is unnecessary here since it is date, inferred
```

```python
# This code is taken with minor modifications from https://seaborn.pydata.org/generated/seaborn.heatmap.html

# Establish size of figure.
plt.figure(figsize = (50,25))

# Get correlation of variables.
corr = df.corr()

# Set up mask to be "True" in the upper triangle.
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True

# Plot our correlation heatmap, while masking the upper triangle to be white.
sns.heatmap(data = corr,    # data to display
            mask = mask,    # mask to restrict what data is displayed
            square = True,  # plot is a square if True
            cmap = 'mako',  # color palette 
            annot = True);  # annotations on plot
```
